```python
import pandas as pd
import folium
```

```python
raw_data = pd.read_csv("C:\\Users\\i00272\\Desktop\\Hour COUNT (Residential Burglary).csv")
test_data = raw_data.copy()
test_data = test_data.dropna(axis=0, how='any')
test_data
```

```python
no_hour = test_data.drop(['HOUR'], axis=1)
no_hour
```

```python
coordinates = [42.3601, -71.0589]

def generateBaseMap(location=coordinates, zoom_start=12):
    base_map = folium.Map(location=coordinates, zoom_start=zoom_start)
    return base_map

base_map = generateBaseMap()
#base_map
```

```python
##create a heat map
from folium.plugins import HeatMap

base = folium.Map(location=coordinates, zoom_start=15)
HeatMap(data=no_hour).add_to(base)
base
```

```python
df_hour_list = []
for hour in test_data.HOUR.sort_values().unique():
    df_hour_list.append(test_data.loc[test_data.HOUR == hour, ['Lat', 'Long', 'Count']].groupby(['Lat', 'Long']).sum().reset_index().values.tolist())

df_hour_list
```

```python
from folium.plugins import HeatMapWithTime

base_map = generateBaseMap(zoom_start=11)
HeatMapWithTime(df_hour_list, radius=15, gradient={0.2: 'blue', 0.4: 'lime', 0.6: 'orange', 1: 'red'}, min_opacity=0.5, max_opacity=0.8, use_local_extrema=True).add_to(base_map)
base_map

## hour 1 = 12:00 AM - 12:59 AM, hour 24 = 11:00 PM - 11:59 PM
```
