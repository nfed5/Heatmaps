```python
import pandas as pd
import folium
```

# Import Data

```python
raw_data = pd.read_csv("C:\\Users\\i00272\\Desktop\\Day of Week COUNT.csv")
test_data = raw_data.copy()
test_data = test_data.dropna(axis=0, how='any')
test_data['Day'] = test_data['Day'].map({'Sunday':1, 'Monday':2, 'Tuesday':3, 'Wednesday':4, 'Thursday':5,'Friday':6,'Saturday':7})
test_data.head()
```

# Base Map

```python
coordinates = [42.3601, -71.0589]

def generateBaseMap(location=coordinates, zoom_start=12):
    base_map = folium.Map(location=coordinates, zoom_start=zoom_start)
    return base_map

base_map = generateBaseMap()
#base_map
```

# Create list of hours

```python
day_list = []
for day in test_data.Day.sort_values().unique():
    day_list.append(test_data.loc[test_data.Day == day, ['Lat', 'Long', 'Count']].groupby(['Lat', 'Long']).sum().reset_index().values.tolist())

#day_list
```

# Heat Map

```python
from folium.plugins import HeatMapWithTime

base_map = generateBaseMap(zoom_start=11)
HeatMapWithTime(day_list, radius=13, gradient={0.2: 'blue', 0.4: 'lime', 0.6: 'orange', 1: 'red'}, min_opacity=0.5, max_opacity=0.8, use_local_extrema=True).add_to(base_map)
base_map

## Sunday = 1, Saturday = 7
```
